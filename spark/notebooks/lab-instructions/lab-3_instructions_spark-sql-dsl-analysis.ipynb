Rédiger les 2 requêtes précédentes avec le DSL de DataFrame.

```python
from pyspark.sql.functions import col, desc, count
```

```python
result_df = trip_df.select(
    "StartTerminal"
).filter(
    col("StartTerminal") == col("EndTerminal")
).join(
    station_df, on=col("StartTerminal") == col("station_id")
).groupBy(
    "StartTerminal","dockcount"
).agg(
    count("*").alias("nombre")
).orderBy(
    desc("nombre")
)
```
